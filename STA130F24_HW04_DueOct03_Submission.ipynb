When a confidence interval overlaps 0, it indicates that you "failed to reject the null hypothesis" because the null hypothesis often assumes no effect or difference. A confidence interval represents a range of probable values for a population parameter. If zero is included, it indicates that no effect is a realistic possibility given the data, implying that there is insufficient evidence to reject the null hypothesis. Bootstrapping Process:

Bootstrapping is a resampling technique where you repeatedly sample with replacement from your data to create multiple "bootstrap samples."
For each bootstrap sample, you calculate the statistic of interest (e.g., mean) and create a distribution of those statistics.
This distribution helps estimate properties like confidence intervals and standard errors without making strong assumptions about the population.
Purpose of Bootstrapping:

Bootstrapping helps estimate the sampling distribution of a statistic, especially when the true distribution is unknown or too complex to calculate analytically.
It can be used to estimate confidence intervals and test hypotheses.
Using Bootstrapping to Assess a Hypothesized Mean:

Given a hypothesized population mean and a sample of size 
𝑛
n, you can use bootstrapping to generate many bootstrap samples and calculate the sample mean for each.
Then, create a confidence interval from the bootstrapped means. If the hypothesized mean falls within this confidence interval, it suggests the hypothesized value is plausible. If it falls outside, the hypothesized mean may not be reasonable.
Confidence Interval Overlapping Zero and the Null Hypothesis:

If a confidence interval includes zero, you "fail to reject the null hypothesis" because zero (no effect or difference) is a plausible value for the population parameter, even if the observed sample mean is not zero.
To reject the null hypothesis, the confidence interval should exclude zero, indicating that the true population parameter is likely different from zero, providing evidence against the null hypothesis.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Load data
data = pd.read_csv('https://raw.githubusercontent.com/NicoLiao/STA130/refs/heads/main/vaccine_data.csv')

# Calculate difference between FinalHealthScore and InitialHealthScore
data['HealthScoreDiff'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Bootstrapping function
def bootstrap(data, n_bootstrap=10000):
    boot_means = []
    for _ in range(n_bootstrap):
        sample = np.random.choice(data['HealthScoreDiff'], size=len(data), replace=True)
        boot_means.append(np.mean(sample))
    return np.array(boot_means)

# Perform bootstrapping
bootstrap_means = bootstrap(data)

# Calculate 95% Confidence Interval
conf_interval = np.percentile(bootstrap_means, [2.5, 97.5])

# Plot the bootstrap distribution
plt.hist(bootstrap_means, bins=30, edgecolor='black')
plt.axvline(conf_interval[0], color='red', linestyle='--', label=f"2.5th percentile: {conf_interval[0]:.2f}")
plt.axvline(conf_interval[1], color='red', linestyle='--', label=f"97.5th percentile: {conf_interval[1]:.2f}")
plt.axvline(np.mean(data['HealthScoreDiff']), color='blue', linestyle='-', label=f"Mean Difference: {np.mean(data['HealthScoreDiff']):.2f}")
plt.title('Bootstrap Distribution of Health Score Differences')
plt.xlabel('Mean Health Score Difference')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Print confidence interval and conclusion
print(f"95% Confidence Interval: {conf_interval}")
if conf_interval[0] > 0:
    print("Reject the null hypothesis. The vaccine appears to be effective.")
else:
    print("Fail to reject the null hypothesis. No sufficient evidence that the vaccine is effective.")
````
