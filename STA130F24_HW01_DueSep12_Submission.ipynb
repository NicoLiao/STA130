````python
import pandas as pd

# Load your dataset
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv')

# Generate a summary for numerical columns
numerical_summary = df.describe()

# Generate a summary for categorical columns
categorical_summary = {}
for column in df.select_dtypes(include=['object']).columns:
    categorical_summary[column] = df[column].value_counts()

# Print the results
print("Numerical Summary:")
print(numerical_summary)

print("\nCategorical Summary:")
for column, counts in categorical_summary.items():
    print(f"\n{column}:")
    print(counts)
````
